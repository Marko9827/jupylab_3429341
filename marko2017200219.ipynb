```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'math score', y = 'reading score', data = df, hue = 'race/ethnicity', palette = 'viridis', s = 200, alpha = 0.5)
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'math score', y = 'reading score', data = df, style = 'parental level of education', hue = 'writing score', size = 'reading score')
```

```python
sns.rugplot(x = 'writing score', data = df, height = 0.5)
```

```python
sns.displot(data = df, x = 'writing score', kde = True)
```

```python
sns.histplot(data = df, x = 'writing score', bins = 100)
```

```python
sns.displot(data = df, x = 'writing score', bins = 20, kde = False, color = 'red', edgecolor = 'black', lw = 4, ls = '--')
```

```python
sns.kdeplot(data = df, x = 'writing score', clip = [200000, 400000])
```

```python
sns.kdeplot(data = df, x = 'writing score', y = 'math score')
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.countplot(x = 'math score', data = df)
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.countplot(x = 'parental level of education', data = df, hue = 'test preparation course', palette = 'Paired')
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'parental level of education', y = 'math score', data = df, hue = 'gender')
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(y = 'parental level of education', x = 'math score', data = df, hue = 'gender')
```

```python
plt.figure(figsize = (12, 6))
sns.violinplot(x = 'parental level of education', y = 'math score', data = df)
```

```python
sns.swarmplot(x = 'math score', data = df, size = 2, y = 'race/ethnicity')
```

```python
sns.swarmplot(y = 'math score', data = df, size = 2, x = 'race/ethnicity', hue = 'gender')
```
